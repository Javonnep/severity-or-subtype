```python
# More masks

mask_cluster_0 = df_ya["cluster"]==0
mask_cluster_1 = df_ya["cluster"]==1
mask_cluster_2 = df_ya["cluster"]==2
```

```python
meta_cols=["SITE_ID","DX_GROUP","AGE_AT_SCAN ","SRS_EDITION", "SRS_VERSION", "SEX", "cluster"]
meta_cols=["SITE_ID","DX_GROUP","SRS_EDITION", "SRS_VERSION", "SEX", "cluster"]

for col in df_ya.drop(meta_cols,axis=1).columns:

    # Get means and std's
    C0_TRAIT = np.array(df_ya[mask_cluster_0 & mask_nan][col])
    C1_TRAIT = np.array(df_ya[mask_cluster_1 & mask_nan][col])
    C2_TRAIT = np.array(df_ya[mask_cluster_2 & mask_nan][col])
    
    ALL_TRAIT = np.array(df[mask_all & mask_nan][col])
    print("="*70)
    print("Means")
    
    print(f"The mean {col} of cluster 0 is {round(np.nanmean(C0_TRAIT),2)} ({round(np.std(C0_TRAIT[~np.isnan(C0_TRAIT)]),2)})")
    print(f"The mean {col} of cluster 1 is {round(np.nanmean(C1_TRAIT),2)} ({round(np.std(C1_TRAIT[~np.isnan(C1_TRAIT)]),2)})")
    print(f"The mean {col} of cluster 2 is {round(np.nanmean(C2_TRAIT),2)} ({round(np.std(C2_TRAIT[~np.isnan(C2_TRAIT)]),2)})")
    print(f"The mean {col} of Allistic subjects is {round(np.nanmean(ALL_TRAIT),2)} ({round(np.std(ALL_TRAIT),2)})")

    # Check for significance
    _, p_value_C0_C1 = stats.ttest_ind(C0_TRAIT[~np.isnan(C0_TRAIT)], C1_TRAIT[~np.isnan(C1_TRAIT)])
    _, p_value_C1_C2 = stats.ttest_ind(C1_TRAIT[~np.isnan(C1_TRAIT)], C2_TRAIT[~np.isnan(C2_TRAIT)])
    _, p_value_C0_C2 = stats.ttest_ind(C0_TRAIT[~np.isnan(C0_TRAIT)], C2_TRAIT[~np.isnan(C2_TRAIT)])
    
    _, p_value_ALL_C0 = stats.ttest_ind(ALL_TRAIT[~np.isnan(ALL_TRAIT)], C0_TRAIT[~np.isnan(C0_TRAIT)])
    _, p_value_ALL_C1 = stats.ttest_ind(ALL_TRAIT[~np.isnan(ALL_TRAIT)], C1_TRAIT[~np.isnan(C1_TRAIT)])
    _, p_value_ALL_C2 = stats.ttest_ind(ALL_TRAIT[~np.isnan(ALL_TRAIT)], C2_TRAIT[~np.isnan(C2_TRAIT)])


    print("="*30)
    print("AUTISTIC vs ALLISTIC")
        
    # if p_value_ALL_C0 < 0.05:
    #     print(f"Difference between ALL and C0 is statistically significant for {col} {round(p_value_ALL_C0,5)}")
    # else:
    #     print(f"Difference between ALL and C0 is NOT statistically significant for {col} {round(p_value_ALL_C0,5)}")

    # if p_value_ALL_C1 < 0.05:
    #     print(f"Difference between ALL and C1 is statistically significant for {col} {round(p_value_ALL_C1,5)}")
    # else:
    #     print(f"Difference between ALL and C1 is NOT statistically significant for {col} {round(p_value_ALL_C1,5)}")

    # if p_value_ALL_C2 < 0.05:
    #     print(f"Difference between ALL and C2 is statistically significant for {col} {round(p_value_ALL_C2,5)}")
    # else:
    #     print(f"Difference between ALL and C2 is NOT statistically significant for {col} {round(p_value_ALL_C2,5)}")

    print("="*30)
    print("INTERGROUP")
    
    if p_value_C0_C1 < 0.05:
        print(f"Difference between C0 and C1 is statistically significant for {col} {round(p_value_C0_C1,5)}")
    else:
        print(f"Difference between C0 and C1 is NOT statistically significant for {col} {round(p_value_C0_C1,5)}")

    if p_value_C1_C2 < 0.05:
        print(f"Difference between C1 and C2 is statistically significant for {col} {round(p_value_C1_C2,5)}")
    else:
        print(f"Difference between C1 and C2 is NOT statistically significant for {col} {round(p_value_C1_C2,5)}")

    if p_value_C0_C2 < 0.05:
        print(f"Difference between C0 and C2 is statistically significant for {col} {round(p_value_C0_C2,5)}")
    else:
        print(f"Difference between C0 and C2 is NOT statistically significant for {col} {round(p_value_C0_C2,5)}")

    print("="*70)

    print()
```
